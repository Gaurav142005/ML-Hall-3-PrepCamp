We import the required modules for the question.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import math
```

# **PLOTTING THE POINTS FOR r = 1 to 10**

```python
# Generating 1000 random points in the given space
x_coor = np.random.rand(1000) * 100
y_coor = np.random.rand(1000) * 100

# Generate 1000 random colors
colors = np.random.rand(1000, 3)

# Generating the coordinate system
for r in range(1, 11):
  fig = plt.figure()
  axis = fig.add_subplot(1, 1, 1)

  axis.grid(which='both') # Adding a grid on both x and y axes

  x_ticks = np.arange(0, 101, r)
  y_ticks = np.arange(0, 101, r)

  axis.set_xticks(x_ticks)
  axis.set_yticks(y_ticks)
  axis.grid(alpha=1, color = "black")

  plt.scatter(x_coor, y_coor, c = colors)

  plt.xlabel("X-Axis")
  plt.ylabel("Y-Axis")
  plt.title("%s x %s Grid" %(r, r))

  plt.show()
  print('')
```

# **Getting the Pairs (Code generalized for d dimensions)**

```python
# Generalising the code for d dimensions
def neighbours():
  d = int(input("Enter the dimension: ")) # Dimension
  r = int(input("Enter the grid size: "))
  n = 1000  # No. of points

  points = [[np.random.rand() * 100 for i in range(d)] for j in range(n)] # Generating the 1000 random points
  grid = [[int(points[i][j] // r) for j in range(d)] for i in range(n)]   # Storing the grid iindex for every point to identify neighbours
  delta = [[[abs(grid[i][k] - grid[j][k]) for k in range(d)] for j in range(i + 1, n)] for i in range(n)] # Calculating the change in grid number

  A = []  # Empty list to store all the pairs

  for i in range(n):  # Traversing through each point
    for j in range(i + 1, n): # Traversing through other points after i (Traversing after i removes duplicate pairs)
      flag = True
      for k in range(d):
        if(delta[i][j - i - 1][k] != 1 and delta[i][j - i - 1][k] != 0):
            flag = False
            break
      if flag == True and i != j:  # Neighbouring Point
        D = 0
        for k in range(d):
          D += (points[i][k] - points[j][k]) ** 2
        D = math.sqrt(D)
        #print(points[i], points[j], D)
        if D < r:
            A.append((points[i], points[j], D))

  return A
```
